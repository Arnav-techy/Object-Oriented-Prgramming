Aggregation

```python
class Customer:
  def __init__(self,name,gender,address):
    self.name=name
    self.gender=gender
    self.address=address


  def edit_profile(self,new_name,new_city,new_pin,new_state):
    self.name=new_name
    self.address.change_address(new_city,new_pin,new_state)

  def __str__(self):
    return f"Name: {self.name}, Gender: {self.gender}, Address: {self.address}"



class Address:
  def __init__(self,city,pin,state):
    self.city=city
    self.pin=pin
    self.state=state

  def change_address(self,new_city,new_pin,new_state):
    self.city=new_city
    self.pin=new_pin
    self.state=new_state

```

Abstraction

```python

from abc import ABC,abstractmethod

class Bankapp(ABC):
  def __init__(self,name,pin):
    self.name=name
    self.pin=pin
    self.amt=0

  def deposit(self,deposit,temp):
    if(self.pin==temp):
      self.amt+=deposit

  def deposit(self,withdraw,temp):
    if(self.pin==temp):
      self.amt-=withdraw

  @abstractmethod
  def signup(self):
    pass

class Mobileapp(Bankapp):
  def signup(self,temp):
    if(self.pin==temp):
      print("Signed In",self.name)
    else: print("Invalid pin")

class Webapp(Bankapp):
  def signup(self,temp):
    if(self.pin==temp):
      print("Signed In",self.name)
    else: print("Invalid pin")
```

```python
acc1=Mobileapp("Arnav",9090)
acc2=Webapp("Suraj",9191)
```

```python
acc1.signup(9191)
acc2.signup(9191)
```
