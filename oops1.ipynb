```python
class Atm:
 def __init__(self):
  self.pin=""
  self.balance=0
  self.menu()

 def menu(self):
  user_input=int(input("""
              how would you like to proceed ?
              1. Enter 1 to create pin
              2. enter 2 to check balance
              3. enter 3 to deposit
              4. enter 4 to withdraw
              5. enter 5 to exit
  """))
  match user_input:
    case 1:
      self.create_pin()
    case 2:
      self.check_balance()
    case 3:
      self.deposit()
    case 4:
      self.withdraw()
    case 5:
      print("bye")
 def create_pin(self):
  self.pin=input("enter your pin")
  print("pin created successfully")

 def deposit(self):
  temp=input("enter your pin")
  if temp == self.pin :
   self.balance+=int(input("amount to be deposited"))
  else:
    print("invalid pin")

 def withdraw(self):
  temp=input("enter your pin")
  if temp == self.pin :
   amount=int(input("enter amount to withdraw"))
   if amount<=self.balance:
    self.balance-=amount
   else:
    print("amount exceeds balance")
  else:
    print("invalid pin")

 def check_balance(self):
  temp=input("enter your pin")
  if temp == self.pin :
    print(self.balance)
  else:
    print("invalid pin")
```

```python
class Fraction:
  def __init__(self,n,d):
    self.num=n
    self.den=d

  def __str__(self):
    return "{}/{}".format(self.num,self.den)

  def __add__(self,other):
    temp_den=self.den*other.den
    temp_num1=self.num*other.den
    temp_num2=other.num*self.den
    sum=temp_num1+temp_num2
    return "{}/{}".format(sum,temp_den)

  def __sub__(self,other):
    temp_den=self.den*other.den
    temp_num1=self.num*other.den
    temp_num2=other.num*self.den
    diff=temp_num1-temp_num2
    return "{}/{}".format(diff,temp_den)

  def __mul__(self,other):
    temp_num=self.num*other.num
    temp_den=self.den*other.den

    return "{}/{}".format(temp_num,temp_den)

  def __truediv__(self,other):
    temp_num=self.num*other.den
    temp_den=self.den*other.num

    return "{}/{}".format(temp_num,temp_den)
```

```python
x=Fraction(7,8)
y=Fraction(4,5)
print(x+y)
print(x-y)
print(x*y)
print(x/y)
```
